```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv('./data/bitcoin_price.csv', sep=',',header=0)
df2 = pd.read_csv('./data/ethereum_price.csv', sep=',',header=0)
df.head()


```

```python
from scipy.stats import probplot
def qq_plot(x):
    probplot(x, dist='norm', plot=plt)
df2 = df2.assign(Mean = (df2['Open'] + df2['Close']) / 2.)
df = df.assign(Mean = (df['Open'] + df['Close']) / 2.)
x=df['Mean']
qq_plot(x[x<10000])
```

```python
from scipy.stats.kde import gaussian_kde
def kde_plot(x):   
    kde = gaussian_kde(x)
    positions = np.linspace(x.min(), x.max())
    smoothed = kde(positions)
    plt.plot(positions, smoothed)
    
def kde_values(x):   
    kde = gaussian_kde(x)
    positions = np.linspace(x.min(), x.max())
    smoothed = kde(positions)
    return positions, smoothed

x=df['Mean']
x=x.dropna()
kde_plot(x)
plt.title("Distribution of all MonthlyIncome data")
plt.show()
```
